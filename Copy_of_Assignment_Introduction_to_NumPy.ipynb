# <u>**Problem 1**</u>

### Construct a 2-D $10*10$ matrix with the following constraints:

*  ####  The element at $[0,0]$ should be 1
*  ####   The element at $[0,9]$ should be 100
* ####   The element at $[9,0]$ should be 50
* ####   The interval between elements in rows should be constant
* ####   The interval between elements in columns should be constant



```
# Your code here
ten__matrix_1_100 = np.linspace(100,145,10,dtype=int)
ten__matrix_9_0 =np.linspace(1,50,10,dtype=int) 


# using for loop 
new_list=[]
for i in range(10):
  # use liner sapce function and take the above variables  and index value as start and end  
  arr=np.linspace(ten__matrix_9_0[i],ten__matrix_1_100[i],10,dtype=int)
  new_list.append(arr)

# convert list into array
twoD_matric=np.asarray(new_list)

# print th array
print(twoD_matric)
```
